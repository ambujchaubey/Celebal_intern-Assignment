```python
import pandas as pd # for data manipulation
import matplotlib.pyplot as plt # for plotting
import numpy as np # for numerical operations
import seaborn as sns

```

```python
# Load the red wine quality dataset
df = pd.read_csv('WineQT.csv', delimiter=',')   #wine quality dataset
print(df.head()) 
print(df.info()) # information about the dataset
print(df.columns)
```

```python
sns.countplot(x='quality',data=df,palette='muted') # Count plot for wine quality
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True,cmap='coolwarm',fmt=".2f") # Heatmap for feature correlation
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
sns.boxplot(x='quality', y='alcohol', data=df, palette='Set2')
plt.title('Alcohol Content vs Wine Quality') # Box plot for alcohol content vs quality
plt.show()
```

```python
selected_features = ['alcohol', 'sulphates', 'citric acid', 'residual sugar', 'quality'] # Selecting features for pairplot
sns.pairplot(df[selected_features], hue='quality', palette='coolwarm')
plt.suptitle('Pairplot of Selected Features by Wine Quality', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='quality', y='volatile acidity', data=df, palette='Spectral') # Swarm plot for volatile acidity vs quality
plt.title('Volatile Acidity vs Wine Quality')
plt.show()
```
